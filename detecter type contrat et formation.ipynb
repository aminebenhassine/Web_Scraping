```python
# libraries & dependencies
import os
import re
import textract
```

```python
# extract raw files content for any files type directory
def directory_rawpdf(directory):
    abs_path = directory + ('/' if directory[-1] != '/' else '')
    files_map = dict()
    for file in os.listdir(directory):
        text = textract.process(abs_path + file).decode('utf-8')
        files_map[file] = re.sub(' +', ' ', text)
    return files_map
```

```python
def detect_contract_type(dictionary):
    import pandas as pd
    import re
    a = pd.DataFrame(columns=["File","Type_Contrat","Formation"])
    patterns_types = [r".*[aA][lL][tT][eE][rR][nN][aA][nN][cC][eE].*",
                      r".*[sS][tT][aA][gG][eE].*",
                      r".*[cC]\.?[dD]\.?[dD].*",
                      r".*[cC]\.?[dD]\.?[iI].*",
                      r".*[cC]\.?[tT]\.?[tT].*",
                      r".*[cC]\.?[uU]\.?[iI].*",
                      r".*[cC]\.?[aA]\.?[eE].*",
                      r".*[V][I][E].*"]
    patterns_formations = [r'.*[bB][aA][cC]\s?\+\s?2.*',
                           r'.*[bB][aA][cC]\s?\+\s?3.*',
                           r'.*[bB][aA][cC]\s?\+\s?4.*',
                           r'.*[bB][aA][cC]\s?\+\s?5.*',
                           r'.*[bB][aA][cC]\s?\+\s?6.*',
                           r'.*[bB][aA][cC]\s?\+\s?7.*',
                           r'.*[iI][nN][gG][eEéÉ][nN][iI][eE][uU][rR].*',
                           r'.*[iI][nN][gG][eEéÉ][nN][iI][eE][rR][iI][eE].*',
                           r'.*[mM][aA][sS][tT][eE][rR].*',
                           r'.*\s[mM]1\s.*',
                           r'.*\s[mM]2\s.*',
                           r'.*[pP][hH][dD].*',
                           r'.*[dD][oO][Cc][tT].*',
                           r'.*[dD]\.?[uU]\.?[tT].*',
                           r'.*[Bb]\.?[tT]\.?[sS].*',
                           r'.*[Bb]\.?[tT]\.?[pP].*',
                           r'.*\s[cC]\.?[aA]\.?[pP]\s.*',
                           r'.*\s[cC]\.?[aA]\.?[pP][aA]\s.*',
                           r'.*[lL][lL][Mm].*',
                           r'.*[lL][iI][cC][eE][nN][cC][eE].*']
    types = ["Alternance","Stage", "CDD","CDI","CTT","CUI","CAE","VIE", "CIE"]
    formations = ["BAC+2","BAC+3","BAC+4","BAC+5","BAC+6","BAC+7","Ingénieur","Ingénieur","Master","M1",
                  "M2", "PhD","PhD","DUT","BTS","BTP","CAP","CAPA","LLM","Licence"]
    h=0
    x = []
    y = []
    t=0
    for k in dictionary.keys():
        s = dictionary.get(k)
        formation = ""
        i = 0
        j=0
        a = a.append({"File":k,"Type_Contrat":'',"Formation":''},ignore_index=True)
        x.append('')
        for p in patterns_types:
            regex = re.compile(p)
            if regex.search(str(s)):
                x[t]=types[i]
                break
            i=i+1
        for f in patterns_formations:
            regex = re.compile(f)
            if regex.search(str(s)):
                formation = formation+' '+formations[j]
            j=j+1
        y.append(formation)
        t = t+1
    for n in range(0,len(x)):
        if x[n]=="":
            x[n]=None 
        if y[n]=="":
            y[n]=None
    a["Formation"]= y
    a["Type_Contrat"]=x
    return a
```
